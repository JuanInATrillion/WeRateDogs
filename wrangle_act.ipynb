```python
Failed_Ids = []
for tweetID in failed_tweetIds:
    Failed_Ids.append(tweetID)
print('There are a total of {} twitter IDs that couldn\'t be accessed'.format(len(Failed_Ids)))
```

**6)** Fix the values of rating_denominator that are not equal to 10

```python
# Find denominators that are not equal to 10, find out why and fix those. 
print(df_enhanced_clean.rating_denominator.value_counts().to_dict(),'\n')

print('Amount of ratings with denominators greater than 10: ',len(df_enhanced_clean[df_enhanced_clean.rating_denominator != 10]))
print('\n')
```

```python
# Capture the indices of rating_denominator that are not 10. 
large_denoms = df_enhanced_clean[df_enhanced_clean.rating_denominator != 10].index.tolist()
print('There is a total of {} rows that need to be fixed'.format(len(large_denoms)))
```

```python
# Find ratios(number/number) in text for the cases where there is not a 10 as a rating_denominator
for i in range(len(large_denoms)): 
    ratios = re.findall(r"\d+/\d+",df_enhanced_clean.text[large_denoms[i]])
    if len(ratios) > 1: 
        df_enhanced_clean.rating_numerator[large_denoms[i]] = int(ratios[1].split('/')[0])
        df_enhanced_clean.rating_denominator[large_denoms[i]] = int(ratios[1].split('/')[1])
    else: 
        if int(ratios[0].split('/')[1]) % 10 == 0:
            den = int(ratios[0].split('/')[1])
            num = int(ratios[0].split('/')[0])/(den/10)
            df_enhanced_clean.rating_denominator[large_denoms[i]] = 10
            df_enhanced_clean.rating_numerator[large_denoms[i]] = num
        else: 
            df_enhanced_clean.rating_numerator[large_denoms[i]] = None
            df_enhanced_clean.rating_denominator[large_denoms[i]] = None
```

**8)** Remove duplicated expanded_urls

```python
# Split the items in expanded_urls column by column. split() method outputs a list of each row split into a list
url_split = df_enhanced_clean.expanded_urls.str.split(',')

# Find if there are any duplicates in the url_split list that contains more than 1 item.
for item in range(0,len(url_split)): 
    df_enhanced_clean.expanded_urls[item] =  list(set(url_split[item])) # convert to set back into a list

```

```python
# Convert each list in expanded_urls column to a string
df_enhanced_clean.expanded_urls = df_enhanced_clean.expanded_urls.map(lambda x: str(x))

# Remove quotes and square brackets to keep only the unique urls
df_enhanced_clean.expanded_urls = df_enhanced_clean.expanded_urls.map(lambda x: x[2:-2])
```
